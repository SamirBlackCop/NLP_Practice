```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
#sentrences
```

```python
sent = ["the glass of milk",
       "the glass of juice",
       "i am a good boy",
       "I am a good developer",
       "understand the meaning of word",
       "your videos are good"]
```

```python
sent
```

```python
voc_size = 10000
```

#### One hot representation

```python
onehot = [one_hot(words, voc_size)for words in sent]
```

```python
print(onehot)
```

#### word embedding Representation

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_lenth = 8
embeded_doc = pad_sequences(onehot, padding = 'pre', maxlen=sent_lenth)
print(embeded_doc)
```

```python
dim=10
```

```python
#adding embedding layer to take featurized representation
```

```python
model = Sequential()
model.add(Embedding(voc_size, 10, input_length= sent_lenth))
model.compile('adam', 'mse')
```

```python
model.summary()
```

```python
print(model.predict(embeded_doc))
```
